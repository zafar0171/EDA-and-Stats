```python
import pandas as pd
df = pd.read_csv("mysql_google_rss.csv",index_col = 0)
df.tail()
```

```python
df.drop_duplicates(['Headline'],inplace = True)
```

```python
sen = []
news = []
org = []

for text in df.Headline:
    x = text.rsplit("|", maxsplit=1)
    
    if len(x) == 2:
        new = x[1].lstrip().rsplit('-', maxsplit=1)
        new[0] = new[0].strip()
        
        if len(new) > 1:
            new[1] = new[1].strip()
            org.append(new[1])
        else:
            org.append('')  # Handle case where there's no second part
        
        sen.append(x[0])
        
        if len(new) > 1 and new[0] == new[1]:
            print(f"{new[0]}, {new[1]} are the same, inserting NaN instead of {new[0]}")
            news.append('NaN')
        else:
            if len(new[0]) > 24:
                print(f"{new[0]} exceeded the character limit")
                news.append('NaN')
            else:
                news.append(new[0])
            
    else:
        e = x[0].lstrip().rsplit('-', maxsplit=1)
        sen.append(e[0])
        news.append('NaN')
        if len(e) > 1:
            org.append(e[1])
        else:
            org.append('')  # Handle case where there's no second part
            
org = [i.strip() for i in org]
sen = [i.strip() for i in sen]
```

```python
new_df = pd.DataFrame({'headline':sen,'news_category':news,'org':org})

new_df
```

```python
import numpy as np
import re

def extract_date(text):
    # Pattern to capture multiple date formats
    pattern = r"\b(?:January|February|March|April|May|June|July|August|September|October|November|December|Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s\d{1,2}(?:,\s?\d{4})?|" \
              r"\b\d{1,2}(?:st|nd|rd|th)?\s(?:of\s)?(?:January|February|March|April|May|June|July|August|September|October|November|December|Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\b|" \
              r"\b\d{1,2}(?:st|nd|rd|th)?\s(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\b|"\
              r"\b(?:Monday|Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday)\b"  
    
    match = re.search(pattern, text, re.IGNORECASE)
    return match.group() if match else np.nan

# Apply the function
filtered_data['date'] = filtered_data['headline'].apply(extract_date)

# Display the result
print(filtered_data)
```

```python

def phrase_df_builder(phrase_val):
    new_dict = {}
    counter = 0 
    for i in filtered_data.headline:
        
        if i.startswith((phrase_val).capitalize()) | i.startswith((phrase_val).title()):
            if 'on' in i.split(':')[0]:
                val = i.split(':')[0].split('on')[-1] +'_' + str(counter)
                new_dict[val] = i.split(':')[-1]
            else:
                val = i.split(':')[0].split(',')[-1] +'_' + str(counter)
                new_dict[val] = i.split(':')[-1]
            counter += 1
    
    
    df = pd.DataFrame({'date':new_dict.keys(),'stocks':new_dict.values()})
        
    return df
```

```python
import pandas as pd
import spacy
from IPython.core.display import display, HTML

# Load SpaCy model
nlp = spacy.load("en_core_web_trf")

# Define entity types to extract
entity_types = ['ORG', 'PERSON', 'GPE', 'DATE', 'CARDINAL']

# Ensure entity columns are initialized with empty lists
for entity in entity_types:
    if entity not in filtered_data.columns:
        filtered_data[entity] = [[] for _ in range(len(filtered_data))]

# Extract entities and update DataFrame
for index, headline in filtered_data['headline'].items():
    # Process the headline
    doc = nlp(headline)
    
    # Render entities for Jupyter visualization
    rendered_html = spacy.displacy.render(doc, style='ent', jupyter=False)
    display(HTML(rendered_html))  # Display rendered entities
    
    # Extract entities and update DataFrame
    for ent in doc.ents:
        if ent.label_ in entity_types:
            filtered_data.at[index, ent.label_].append(ent.text)

# Display updated DataFrame
print(filtered_data)
```

```python
import pandas as pd
import spacy

# Load SpaCy model
nlp = spacy.load("en_core_web_trf")

# Assuming filtered_data is already defined and contains a 'headline' column
# Initialize lists for each entity type
org_list = []
person_list = []
gpe_list = []
date_list = []
cardinal_list = []

# Extract entities and append them to respective lists
for index, headline in filtered_data['headline'].items():
    # Process the headline
    doc = nlp(headline)
    
    # Initialize temporary lists for the current row
    orgs = []
    persons = []
    gpes = []
    dates = []
    cardinals = []
    
    # Extract entities and classify them into respective lists
    for ent in doc.ents:
        if ent.label_ == 'ORG':
            orgs.append(ent.text)
        elif ent.label_ == 'PERSON':
            persons.append(ent.text)
        elif ent.label_ == 'GPE':
            gpes.append(ent.text)
        elif ent.label_ == 'DATE':
            dates.append(ent.text)
        elif ent.label_ == 'CARDINAL':
            cardinals.append(ent.text)
    
    # Append the current row's entity lists to the main lists
    org_list.append(orgs)
    person_list.append(persons)
    gpe_list.append(gpes)
    date_list.append(dates)
    cardinal_list.append(cardinals)

# Assign the lists to the DataFrame columns
filtered_data['ORG'] = org_list
filtered_data['PERSON'] = person_list
filtered_data['GPE'] = gpe_list
filtered_data['DATE'] = date_list
filtered_data['CARDINAL'] = cardinal_list

# Display updated DataFrame
print(filtered_data)
```
